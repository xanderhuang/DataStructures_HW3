```python
def calculateRatio(l):
  ratios = []
  for i in range(len(l)):
      if l[i]!=0 and i<len(l)-1:
        ratio = l[i+1] / l[i] 
        ratios.append(ratio)
  return round(sum(ratios)/len(ratios), 2)

for i in range(len(insert_df['HashTable'])):
    if i!=0 and insert_df['HashTable'][i] == float(0):
      insert_df['HashTable'][i] = insert_df['HashTable'][i-1] * calculateRatio(HashTable)

for i in range(len(insert_df['BTree'])):
    if i!=0 and insert_df['BTree'][i] == float(0):
      insert_df['BTree'][i] = insert_df['BTree'][i-1] * calculateRatio(BTree)

for i in range(len(insert_df['Treap'])):
    if i!=0 and insert_df['Treap'][i] == float(0):
      insert_df['Treap'][i] = insert_df['Treap'][i-1] * calculateRatio(Treap)

for i in range(len(insert_df['SkipList'])):
    if i!=0 and insert_df['SkipList'][i] == float(0):
      insert_df['SkipList'][i] = insert_df['SkipList'][i-1] * calculateRatio(SkipList)

pd.set_option('display.float_format', lambda x: '%.2f' % x)
insert_df
```
